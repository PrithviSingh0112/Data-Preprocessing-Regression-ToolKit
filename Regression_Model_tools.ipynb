Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Data For Linear Regression

```python
dataset_2 = pd.read_csv('Salary_Data.csv')
x_1 = dataset_2.iloc[:, :-1].values
y_1 = dataset_2.iloc[:, -1].values
```

Splitting the Dataset into Training set and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_1, y_1, test_size = 1/3, random_state = 0)
```

Training Simple linear Regression model on training set

```python
from sklearn.linear_model import LinearRegression
simple_regressor = LinearRegression()
simple_regressor.fit(x_train, y_train)
```

Predicting Test set values

```python
y_pred = simple_regressor.predict(x_test)
```

Visualizing train set

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, simple_regressor.predict(x_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualising test set

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, y_pred, color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Importing Data

```python
dataset_1 = pd.read_csv('50_Startups.csv')
x_2 = dataset_1.iloc[: , :-1].values
y_2 = dataset_1.iloc[:,-1].values
```

```python
print(x_2)
```

Encoding Categorical Data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])] , remainder='passthrough')
x_2 = np.array(ct.fit_transform(x_2))
```

```python
print(x_2)
```

Splitting Train and Test Data

```python
from  sklearn.model_selection import train_test_split
x_train , x_test,y_train,y_test = train_test_split(x_2,y_2,test_size = 0.2 , random_state=0)
```

Train Multiple Linear regression Model

```python
from sklearn.linear_model import LinearRegression
multiple_regressor = LinearRegression()
multiple_regressor.fit(x_train , y_train)
```

Test Predictions After training the Model

```python
y_pred = multiple_regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Importing Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[: ,1:-1].values
y = dataset.iloc[: , -1].values
```

```python
print(x)
```

```python
print(y)
```

 Training the Decision Tree Regression model on the Dataset

```python
from sklearn.tree import DecisionTreeRegressor
tree_regressor= DecisionTreeRegressor(random_state=0)
tree_regressor.fit(x, y)
```

 Predict Results

```python
tree_regressor.predict([[6.5]])
```

Visualising the Decision Tree Regression results (higher resolution)

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, tree_regressor.predict(x_grid), color = 'blue')
plt.title('(Decision tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Train Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
forest_regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
forest_regressor.fit(x,y)
```

Predict Value

```python
forest_regressor.predict([[6.5]])
```

Visualize prediction Of Random Forest

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, forest_regressor.predict(x_grid), color = 'blue')
plt.title('(Random forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Support Vector Regression (SVR)

```python
y = y.reshape(len(y),1)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x = sc_x.fit_transform(x)
y = sc_y.fit_transform(y)
```

```python
print(x)
```

```python
print(y)
```

Training the SVR model on the whole dataset

```python
from sklearn.svm import SVR
SVR_regressor = SVR(kernel = 'rbf')
SVR_regressor.fit(x, y)
```

Predicting a new result

```python
sc_y.inverse_transform(SVR_regressor.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

```python
x_grid = np.arange(min(sc_x.inverse_transform(x)), max(sc_x.inverse_transform(x)), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(sc_x.inverse_transform(x), sc_y.inverse_transform(y), color = 'red')
plt.plot(x_grid, sc_y.inverse_transform(SVR_regressor.predict(sc_x.transform(x_grid)).reshape(-1,1)), color = 'blue')
plt.title('Truth or Bluff (SVR)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
